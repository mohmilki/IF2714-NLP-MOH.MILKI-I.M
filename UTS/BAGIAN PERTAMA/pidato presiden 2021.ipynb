Memanggil library stopwords (berfungsi untuk menghilangkan kata-kata yang tidak bermakna), word_tokenize (berfungsi untuk menjadikan sebuah kalimat menjadi list kata) dan matplotlib.pyplot untuk plotting kata-kata yang sering muncul pada isi pidato

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
```

Melihat isi pidato jokowi tahun 2021

```python
f = open('pidato 2021.txt')
isi_pidato = f.read()
```

```python
isi_pidato
```

Mengubah isi pidato menjadi huruf kecil

```python
isi_pidato_lower_case = isi_pidato.lower()
```

Melihat isi pidato presiden yang seluruhnya telah menjadi huruf kecil

```python
isi_pidato_lower_case
```

Mengubah seluruh isi pidato kedalam list kata dengan word_tokenize

```python
word_tokens = word_tokenize(isi_pidato_lower_case)
```

```python
word_tokens
```

Mengisiasi seluruh kata bahasa Indonesia yang tidak mempunyai makna kedalam variabel stop_words

```python
stop_words = set(stopwords.words('indonesian'))
```

Memisahkan isi pidato yang tidak ada stopwords ke dalam variabel word_tokens_no_stopwords

```python
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
```

**CLEANING DATA**

Menghapus seluruh simbol pada isi pidato presiden jokowi dan menyimpannya di variabel filtered_symbols

```python
unuseful_symbols = "',.()\"\``\’“”';!?:'"
filtered_symbols = []
for w in word_tokens_no_stopwords:
  if w not in unuseful_symbols:
    filtered_symbols.append(w)
```

```python
filtered_symbols
```

Melihat frekuensi kata (yang sering muncul) dengan nltk.FreqDist yang disimpan pada variabel freq_kata_symbols setelah seluruh simbol dihilangkan

```python
freq_kata_symbols = nltk.FreqDist(filtered_symbols)
```

```python
freq_kata_symbols.most_common()
```

Menghilangkan kata-kata pada isi pidato yang sering muncul namun tidak mempunyai arti (dalam segi konteks) selain dari stopwords.

```python
unuseful_words = ['hormati','republik','bangsa','ketua','menghadapi']
filtered_words = []
for w in filtered_symbols:
  if w not in unuseful_words:
    filtered_words.append(w)
```

Melihat frekuensi kata (yang sering muncul) dengan nltk.FreqDist yang disimpan pada variabel freq_kata setelah isi pidato di cleaning

```python
freq_kata = nltk.FreqDist(filtered_words)
```

Melihat 10 kata yang sering muncul

```python
freq_kata.most_common(10)
```

Melihat 5 kata yang sering muncul, ditampilkan dengan plotting

```python
freq_kata.plot(5)
```

Melihat 10 kata yang sering muncul, ditampilkan dengan plotting

```python
freq_kata.plot(10)
```
