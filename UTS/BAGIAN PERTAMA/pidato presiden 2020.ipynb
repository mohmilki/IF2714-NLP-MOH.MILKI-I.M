Memanggil data isi pidato presiden tahun 2020

```python
f = open('pidato 2020.txt')
isi_pidato = f.read()
```

Melihat variabel isi_pidato yang berisi pidato presiden jokowi tahun 2019

```python
isi_pidato
```

Memanggil library stopwords (berfungsi untuk menghilangkan kata-kata yang tidak bermakna) dan word_tokenize (berfungsi untuk menjadikan sebuah kalimat menjadi list kata)

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

Mengubah seluruh isi pidato menjadi huruf kecil

```python
isi_pidato_lower_case = isi_pidato.lower()
```

```python
isi_pidato_lower_case
```

Mengubah isi pidato menjadi list kata

```python
word_tokens = word_tokenize(isi_pidato_lower_case)
```

```python
word_tokens
```

Menginisiasi daftar kata yang tidak memiliki arti ke dalam variabel stop_wirds dan Menghilangkan simbol-simbol yang tidak berguna pada isi pidato

```python
stop_words = set(stopwords.words('indonesian'))
```

```python
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
```

```python
unuseful_symbols = "',.()\"\``\’';!?:'"
filtered_symbols = []
for w in word_tokens_no_stopwords:
  if w not in unuseful_symbols:
    filtered_symbols.append(w)
```

```python
filtered_symbols
```

Melihat jumlah kata yang sering muncul pada isi pidato jokowi setelah seluruh simbol yang tidak perlu dihilangkan

```python
freq_kata_symbols = nltk.FreqDist(filtered_symbols)
```

```python
freq_kata_symbols.most_common()
```

Menghilangkan kata-kata yang sering muncul namun tidak mempunyai arti (atau tidak sesuai konteks) pada isi pidato jokowi

```python
unuseful_words = ['hormati','republik','se-bangsa','se-tanah','air','ketua','saudara-saudara','2020','cepat','kemajuan','anggota','kecepatan']
filtered_words = []
for w in filtered_symbols:
  if w not in unuseful_words:
    filtered_words.append(w)
```

Melihat frekuensi kata-kata yang sering muncul pada isi pidato presiden jokowi setelah cleaning data

```python
freq_kata = nltk.FreqDist(filtered_words)
```

```python
freq_kata.most_common()
```
