Memanggil data isi pidato presiden tahun 2019

```python
f = open('./BAGIAN PERTAMA/pidato 2019.txt')
isi_pidato = f.read()
```

Melihat variabel isi_pidato yang berisi pidato presiden jokowi tahun 2019

Mengubah seluruh isi pidato menjadi huruf kecil

```python
lower_case_isi_pidato = isi_pidato.lower()
```

```python
lower_case_isi_pidato
```

Mengubah isi pidato menjadi list kata

```python
word_tokens = word_tokenize(lower_case_isi_pidato)
```

```python
word_tokens
```

Menginisiasi variabel stop_words yang berisi seluruh kata-kata dalam bahasa Indonesia yang tidak mempunyai makna

```python
stop_words = set(stopwords.words('indonesian'))
```

Menghilangkan kata-kata yang tidak berguna dan menyimpannya ke variabel word_tokens_no_stopwords

```python
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
```

Meliihat seluruh isi varibel word_tokens_no_stopwords

```python
word_tokens_no_stopwords
```

**PROSES CLEANING**

Menghilangkan simbol-simbol yang tidak berguna pada isi pidato

```python
unuseful_symbols = "',.()\"\``\’';!?:'"
filtered_symbols = []
for w in word_tokens_no_stopwords:
  if w not in unuseful_symbols:
    filtered_symbols.append(w)
```

```python
filtered_symbols
```

Menghilangkan kata-kata yang tidak mempunyai makna terkait isi pidato (yang tidak dapat dihilangkan oleh stopwords) yang mengandung jumlah kata yang banyak.

```python
unuseful_words = ['butuh','hormati','lompatan','diukur','menghadapi']
filtered_words = []
for w in filtered_symbols:
  if w not in unuseful_words:
    filtered_words.append(w)
```

Menginisiasi frekuensi isi pidato yang telah di cleaning dengan nltk.FreqDist dan menyimpannya di variabel freq_data

```python
freq_kata = nltk.FreqDist(filtered_words)
```

Melihat seluruh kata-kata yang sering muncul pada isi pidato presiden jokowi

```python
freq_kata.most_common()
```
