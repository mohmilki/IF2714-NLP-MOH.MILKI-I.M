Membuat fungsi dengan nama clean_text yang berfungsi untuk cleaning data

```python
def clean_text(text):
  delete_dict ={sp_character: '' for sp_character in string.punctuation}
  delete_dict[' '] = ' '
  table = str.maketrans(delete_dict)
  text1 = text.translate(table)
  textArr = text1.split()
  text2 = ' '.join([w for w in textArr if (not w.isdigit() and (not w.isdigit() and len(w)>3))])

  return text2.lower()
```

Import library nltk (stopwords) untuk menghilangkan kata-kata yang tidak penting

```python
import nltk
nltk.download('stopwords')
```

```python
df.dropna(axis = 0, how = 'any', inplace=True)
```

```python
df.head()
```

```python
df['Text'] = df['Text'].apply(clean_text)
```

Membuat variabel yang bernama stopwords yang berisi kata-kata yang tidak mempunyai makna dalam bahasa Inggris

```python
stopwords = stopwords.words('english')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

def remove_stopwords(text):
  textArr = text.split(' ')
  rem_text = " ".join([i for i in textArr if i not in stop_words])
  return rem_text

df['Text'] = df['Text'].apply(remove_stopwords)
```

```python
df['Text'].head()
```

Proses pengelompokkan fitur Text menjadi beberapa topic (tidak disetting jumlahnya, sehingga jumlah topic diberikan secara default)

```python
model = BERTopic(verbose=True)

docs = df.Text.to_list()
topics, probabilities = model.fit_transform(docs)
```
