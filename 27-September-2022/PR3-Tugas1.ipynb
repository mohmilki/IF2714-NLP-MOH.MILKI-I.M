```python
# Read Data (Baca data)
amz_review = pd.read_csv('imdb_labelled.txt', sep='\t', names=['review', 'label'])

# Take a look at the data
amz_review.head()
```

```python
# get the dataset information (memperoleh informasi dari dataset)
amz_review.info()
```

```python
amz_review.head()
```

```python
# Check the label distribution (melihat distribusi/persebaran label)
amz_review['label'].value_counts()
```

```python
# Get sentiment score for each review (Memperoleh nilai sentiment untuk setiap review)

vader_sentiment = SentimentIntensityAnalyzer()
amz_review['scores_VADER'] = amz_review['review'].apply(lambda s: vader_sentiment.polarity_scores(s)['compound'])

# Predict sentiment label for each review (Memprediksi label sentiment untuk setiap review)
amz_review['pred_VADER'] = amz_review['scores_VADER'].apply(lambda x: 1 if x >=0 else 0)
amz_review.head()
```

```python
# compare Actual and Predicted
accuracy_score(amz_review['label'],amz_review['pred_VADER'])
```
