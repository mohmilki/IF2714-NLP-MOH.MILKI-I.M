```python
# Read Data
amz_review = pd.read_table('a1_RestaurantReviews_HistoricDump.tsv')

# Take a look at the data
amz_review.head()
```

```python
# Get the dataset information
amz_review.info()
```

```python
amz_review.head()
```

```python
# Check the label distribution
amz_review['Liked'].value_counts()
```

```python
# Get sentiment score for each review

vader_sentiment = SentimentIntensityAnalyzer()
amz_review['score_VADER'] = amz_review['Review'].apply(lambda s: vader_sentiment.polarity_scores(s)['compound'])

# Predict sentiment label for each review
amz_review['pred_VADER'] = amz_review['score_VADER'].apply(lambda x: 1 if x>=0 else 0)
amz_review.head()
```

```python
# Compare Actual and Predicted
accuracy_score(amz_review['Liked'],amz_review['pred_VADER'])
```
