```python
f = open('tugas1.txt')
isi_berita = f.read()
print(isi_berita)
```

```python
word_tokens =  word_tokenize(isi_berita)
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
import matplotlib.pyplot as plt
```

```python
freq_kata_1 = nltk.FreqDist(word_tokens)
freq_kata_2 = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata_1.most_common(25))
print()
print(freq_kata_2.most_common(25))
```

```python
freq_kata_1.plot(25)
freq_kata_2.plot(25)
plt.show()
```

```python
forbidden = "',.()\"\``\''"
filtered_no_forbidden_and_stopwords = []
for w in word_tokens_no_stopwords:
  if w not in forbidden:
    filtered_no_forbidden_and_stopwords.append(w)
```

```python
print(filtered_no_forbidden_and_stopwords)
```

```python
freq_kata_3 = nltk.FreqDist(filtered_no_forbidden_and_stopwords)
```

```python
freq_kata_3.plot(30)
```

```python
print(freq_kata_3.most_common(25))
```
